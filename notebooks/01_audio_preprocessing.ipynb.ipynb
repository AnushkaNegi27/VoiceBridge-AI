```python
import os
import subprocess
```

```python
RAW_DIR = "../data/common_voice_hindi/clips"
OUT_DIR = "../data/common_voice_hindi/processed_clips"

os.makedirs(OUT_DIR, exist_ok=True)

print("Raw dir exists:", os.path.exists(RAW_DIR))
print("Output dir created:", os.path.exists(OUT_DIR))
```

```python
count = 0

for file in os.listdir(RAW_DIR):
    if file.endswith((".mp3", ".wav")):
        inp = os.path.join(RAW_DIR, file)
        out = os.path.join(
            OUT_DIR,
            os.path.splitext(file)[0] + ".wav"
        )

        subprocess.run(
            ["ffmpeg", "-y", "-i", inp, "-ar", "16000", "-ac", "1", out],
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL
        )

        count += 1

print("Total converted files:", count)
```

```python
import pandas as pd
```

```python
tsv_path = "../data/common_voice_hindi/train.tsv"
df = pd.read_csv(tsv_path, sep="\t")

print("Total rows in train.tsv:", len(df))
df.head()
```

```python
df_clean = df[["path", "sentence", "client_id"]]

print("Columns kept:", df_clean.columns.tolist())
df_clean.head()
```

```python
BASE_AUDIO_DIR = "../data/common_voice_hindi/processed_clips"

df_clean["file_path"] = df_clean["path"].apply(
    lambda x: os.path.join(
        BASE_AUDIO_DIR,
        os.path.splitext(x)[0] + ".wav"
    )
)

df_clean.head()
```

```python
missing = df_clean[~df_clean["file_path"].apply(os.path.exists)]

print("Missing audio files:", len(missing))
```

```python
import librosa
```

```python
def get_duration(path):
    try:
        return librosa.get_duration(path=path)
    except:
        return None

df_clean["duration"] = df_clean["file_path"].apply(get_duration)

df_clean.head()
```

```python
print("Missing durations:", df_clean["duration"].isna().sum())
print("Average duration (sec):", df_clean["duration"].mean())
```

```python
metadata = pd.DataFrame({
    "file_path": df_clean["file_path"],
    "duration": df_clean["duration"],
    "language": "hi",
    "speaker_id": df_clean["client_id"],
    "transcript": df_clean["sentence"],
    "quality_flag": "good"
})

metadata.head()
```

```python
metadata_path = "../data/common_voice_hindi/metadata.csv"
metadata.to_csv(metadata_path, index=False)

print("Metadata saved at:", metadata_path)
print("Total metadata rows:", len(metadata))
```

```python
total_duration = metadata["duration"].sum()
avg_duration = metadata["duration"].mean()
num_clips = len(metadata)

print("Total clips:", num_clips)
print("Total Hindi audio duration (minutes):", total_duration / 60)
print("Average clip length (seconds):", avg_duration)
```

```python
from pydub import AudioSegment
from pydub.effects import normalize
import os
```

```python
AUDIO_DIR = "../data/common_voice_hindi/processed_clips"

for file in os.listdir(AUDIO_DIR):
    if file.endswith(".wav"):
        path = os.path.join(AUDIO_DIR, file)
        audio = AudioSegment.from_wav(path)
        audio = normalize(audio)
        audio.export(path, format="wav")

print("Audio normalization completed")
```

```python
import librosa
import soundfile as sf
import os
```

```python
AUDIO_DIR = "../data/common_voice_hindi/processed_clips"

for file in os.listdir(AUDIO_DIR):
    if file.endswith(".wav"):
        path = os.path.join(AUDIO_DIR, file)

        y, sr = librosa.load(path, sr=16000)
        y_trimmed, _ = librosa.effects.trim(y, top_db=20)

        sf.write(path, y_trimmed, sr)

print("Silence trimming completed")
```

```python
import re
```

```python
def clean_text(text):
    text = str(text)
    text = text.replace("\n", " ")
    text = re.sub(r"\s+", " ", text)
    return text.strip()

metadata["transcript"] = metadata["transcript"].apply(clean_text)

print("Transcript cleaning completed")
```

```python
total_duration_sec = metadata["duration"].sum()
total_duration_hr = total_duration_sec / 3600

num_clips = len(metadata)
avg_clip_length = metadata["duration"].mean()

print("----- HINDI DATASET STATS -----")
print("Number of clips:", num_clips)
print("Total duration (seconds):", round(total_duration_sec, 2))
print("Total duration (hours):", round(total_duration_hr, 2))
print("Average clip length (seconds):", round(avg_clip_length, 2))
```

```python
required_columns = [
    "file_path",
    "duration",
    "language",
    "speaker_id",
    "transcript",
    "quality_flag"
]

print("Missing columns:",
      set(required_columns) - set(metadata.columns))

print("Total metadata rows:", len(metadata))
metadata.head()
```

```python
metadata_path = "../data/common_voice_hindi/metadata.csv"
metadata.to_csv(metadata_path, index=False)

print("Final metadata saved at:", metadata_path)
```

English Dataset

```python
RAW_DIR = "../data/english/recordings"
OUT_DIR = "../data/english/processed_clips"

os.makedirs(OUT_DIR, exist_ok=True)

count = 0

for file in os.listdir(RAW_DIR):
    if file.lower().endswith((".mp3", ".wav")):
        inp = os.path.join(RAW_DIR, file)
        out = os.path.join(
            OUT_DIR,
            os.path.splitext(file)[0] + ".wav"
        )

        subprocess.run(
            ["ffmpeg", "-y", "-i", inp, "-ar", "16000", "-ac", "1", out],
            stdout=subprocess.DEVNULL,
            stderr=subprocess.DEVNULL
        )

        count += 1

print("Total English files converted:", count)
```

```python
with open("../data/english/reading-passage.txt", "r", encoding="utf-8") as f:
    EN_TRANSCRIPT = f.read().strip()

print("Transcript loaded:")
print(EN_TRANSCRIPT[:120], "...")
```

```python
AUDIO_DIR = "../data/english/processed_clips"

rows = []

for file in os.listdir(AUDIO_DIR):
    if file.endswith(".wav"):
        path = os.path.join(AUDIO_DIR, file)

        try:
            duration = librosa.get_duration(path=path)
        except:
            continue

        rows.append({
            "file_path": path,
            "duration": round(duration, 3),
            "language": "en",
            "speaker_id": os.path.splitext(file)[0],
            "transcript": EN_TRANSCRIPT,
            "quality_flag": "good"
        })

metadata_en = pd.DataFrame(rows)

print("Total English metadata rows:", len(metadata_en))
metadata_en.head()
```

```python
metadata_en_path = "../data/english/metadata.csv"
metadata_en.to_csv(metadata_en_path, index=False)

print("English metadata saved at:", metadata_en_path)
```

```python
# Load Hindi & English metadata
hi_meta = pd.read_csv("../data/common_voice_hindi/metadata.csv")
en_meta = pd.read_csv("../data/english/metadata.csv")

print("Hindi rows:", len(hi_meta))
print("English rows:", len(en_meta))

# Merge
final_metadata = pd.concat([hi_meta, en_meta], ignore_index=True)

print("Total merged rows:", len(final_metadata))

final_metadata.head()
```

```python
final_path = "../data/final_metadata.csv"
final_metadata.to_csv(final_path, index=False)

print("Final merged metadata saved at:", final_path)
```
